### **Belajar Supervised Learning ~ Regression**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

#### **Algoritma Linear Regression**

```python
x = np.array([
    [2, 3, 0],    # 2GB i3 Integrated
    [4, 5, 0],    # 4GB i5 Integrated
    [8, 5, 1],    # 8GB i5 RTX 2050
    [8, 7, 1],    # 8GB i7 RTX 2050
    [12,7,1],     # 12GB i7 RTX 2050
    [16,7,2],     # 16GB i7 RTX 3050
    [16,9,2],     # 16GB i9 RTX 3050
    [16,9,3],     # 16GB i9 RTX 4060
    [24,7,3],     # 24GB i7 RTX 4060
    [24,9,4],     # 24GB i9 RTX 5070
    [32,9,4],     # 32GB i9 RTX 5070
])
y = np.array([
    1500,  # 2GB i3 Integrated
    3000,  # 4GB i5 Integrated
    6500,  # 8GB i5 RTX 2050
    8500,  # 8GB i7 RTX 2050
    9500,  # 12GB i7 RTX 2050
    12500, # 16GB i7 RTX 3050
    14500, # 16GB i9 RTX 3050
    16500, # 16GB i9 RTX 4060
    18500, # 24GB i7 RTX 4060
    21000, # 24GB i9 RTX 5070
    23500, # 32GB i9 RTX 5070
])
ram = x[:, 0]
processor = x[:, 1]

data = {
    "ram": ram,
    "processor": processor,
    "price": y
}

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="ram", y="price", hue="processor", s=100, palette="viridis")
plt.title("Laptop Price based on RAM and Processor")
plt.xlabel("RAM (GB)")
plt.ylabel("Price (in thousands)")
plt.legend(title="Processor Intel i")
plt.grid(True)
plt.show()
```

```python
model = LinearRegression()
model.fit(x, y)
print("Intercept b0 (when x = 0):", model.intercept_)
print("Coefficient b1 ram (when x[0] = 1):", model.coef_[0])
print("Coefficient b1 processor (when x[1] = 1):", model.coef_[1])
print("Coefficient b1 VGA (when x[1] = 1):", model.coef_[2])
```
